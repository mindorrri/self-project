```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import matplotlib
```

```python
df = pd.read_csv(r"C:\Users\yms10\OneDrive\문서\spending.csv", thousands = ',')
```

```python
category_total = df.groupby('category')['cost'].sum()
category_total
```

```python
category_total.plot(kind = 'bar', color = 'skyblue')
```

```python
plt.title("Total spending by category")
plt.ylabel("cost")
plt.xlabel("category")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
category_total = df.groupby('category')['cost'].sum()
category_total.plot(kind = 'pie', autopct = '%1.1f%%') #원그래프로 나타내기
```

```python
#요일별 소비 패턴
df["day"] = df['date'].dt.day_name()
weekday_total = df.groupby("day")['cost'].sum()
weekday_total.plot(kind='bar')
```
